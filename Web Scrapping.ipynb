```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36'}
```

```python
response=requests.get('https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating',headers=headers).text
```

```python
response
```

```python

soup = BeautifulSoup(response, 'lxml')
```

```python
print(soup.prettify())
```

```python
data=[]
for i in soup.find_all('h3',class_='lister-item-header'):
    data.append(i.text.strip())
```

```python
data
```

```python
title_list = []
year_list = []

for entry in data:
    entry = entry.strip()  # Remove leading/trailing whitespaces
    split_data = entry.split('\n')
    title = split_data[1]
    year = int(''.join(filter(str.isdigit, split_data[2])))  # Extract digits from the split data
    title_list.append(title)
    year_list.append(year)
    
```

```python
rating=[]
for i in soup.find_all('strong'):
    rating.append(i.text.strip())
```

```python
rating=rating[2:]
```

```python

print(title_list)
print(year_list)
print(rating)
```

```python
runtime=[]
for i in soup.find_all('span',class_='runtime'):
    runtime.append(i.text.strip())
```

```python
genre=[]
for i in soup.find_all('span',class_='genre'):
    genre.append(i.text.strip())
```

```python
Director=[]
Stars=[]
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-advanced'})
for store in movie_data:
    cast = store.find("p", class_ = '')
    cast = cast.text.replace('\n', '').split('|')
    cast = [x.strip() for x in cast]
    cast = [cast[i].replace(j, "") for i,j in enumerate(["Director:", "Stars:"])]
    Director.append(cast[0])
    Stars.append([x.strip() for x in cast[1].split(",")])
```

```python
overview = []

for element in soup.findAll('p', class_='text-muted'):
    text = element.contents[0].strip()
    if text:
        overview.append(text)

print(overview)
```

```python
data=pd.DataFrame({'Title':title_list,'Year':year_list,'Rating':rating,'Run Time':runtime,'Genre':genre,'Director':Director,'Stars':Stars,'Overview':overview})
```

```python
data
```

```python
data.to_csv('imdb_movies_data.csv', index=False)
```

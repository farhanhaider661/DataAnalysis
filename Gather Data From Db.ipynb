```python
import mysql.connector
```

```python
connection = mysql.connector.connect(
    host='127.0.0.1',
    user='root',
    password='1234',
    database='Movies'
)
```

```python
cursor = connection.cursor()
```

```python
cursor.execute("SELECT * FROM pop_movies")
```

```python
results = cursor.fetchall()
```

```python
results
```

```python
Title=[]
Popularity=[]
vote_average=[]
vote_count=[]
release_date=[]
for i in results:
    Title.append(i[1])
    Popularity.append(i[2])
    vote_average.append(i[3])
    vote_count.append(i[4])
    release_date.append(i[5])
```

```python
import pandas as pd
```

```python
data=pd.DataFrame({'Title':Title,'Popularity':Popularity,'vote_average':vote_average,'vote_count':vote_count,'release_date':release_date})
data
```

```python
data.to_csv('movies_db.csv', index=False)
```
